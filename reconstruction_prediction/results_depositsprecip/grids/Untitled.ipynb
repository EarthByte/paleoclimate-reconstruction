```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
names=["lat","lon","precip_actual","precip_mean","precip_low","precip_high","coal_mean","evap_mean",
 "glacial_mean","coal_low","evap_low","glacial_low","coal_high","evap_high","glacial_high",
 "coal_actual","evap_actual","glacial_actual","1","2","3","4","5","6"]
```

```python
df=pd.read_csv("../era249results.csv",names=names)
for n in names[2:-6]:
    print(n)
    df2=pd.read_csv("249Ma_"+n+".xyz",names=["lon","lat",n],sep='\s+')
    df3=pd.read_csv("249Ma_"+n+".block",names=["lon","lat",n],sep='\s+')

    plt.scatter(df.lon,df.lat,c=df[n],s=20)
    plt.show()
    plt.scatter(df2.lon,df2.lat,c=df2[n],s=20)
    plt.show()
    plt.scatter(df3.lon,df3.lat,c=df3[n],s=20)
    plt.show()
```
